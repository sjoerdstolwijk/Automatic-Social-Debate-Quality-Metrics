```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#get the files

path = r"~\OneDrive - UvA\Project\Project public sphere\ICA graphs\ICA graphs Monte Carlo\CSV"

dimensions = ['outcome_conservative','outcome_general_incivility','outcome_interactivity','outcome_liberal',
              'outcome_narrow_incivility','outcome_rationality']

all_dimensions = pd.DataFrame()

for dimension in dimensions:
    file = pd.read_csv(f"{path}\{dimension}.csv")

    display(file.head(5))

    all_dimensions = pd.concat([all_dimensions,file])
```

```python
#plot violin
plt.figure(figsize=(15,8))
sns.violinplot(y = all_dimensions['F1_score'], x = all_dimensions['Variable'])
plt.title('Macro F1 score for each dimension').set_fontsize(15)
plt.xlabel('Variable')
plt.ylabel('F1 Score')
```

```python
#plot line plot f1 score
sns.set_style("whitegrid")
plt.figure(figsize=(20,15))
sns.lineplot(x = all_dimensions['Sample Size'], y = all_dimensions['F1_score'], hue=all_dimensions['Variable'], err_style='bars',
             linewidth = 1.5, style=all_dimensions['Variable'])
plt.ylim(0.2,0.7)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.title('Macro F1 score for each dimension across sample sizes').set_fontsize(15)
plt.xlabel('Sample Size')
plt.ylabel('Macro F1 Score')
```
